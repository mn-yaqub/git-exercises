# Coding style - evaluated exercise

For this exercise, you will need the `Ruff` extension installed and configured in Visual Studio Code.

After installing `Ruff` extension, it can be configured by:

- Pressing `Ctrl + Shift + P` in Visual Studio Code.
- Type `Preferences: Open Settings (JSON)`.
- Insert the three lines into the JSON file:
```json
{
  "ruff.format.preview": true,
  "ruff.lint.preview": true,
  "ruff.lint.select": ["E", "F", "W", "I", "N"]
}
```
This will enable a subset of coding style rules that `Ruff` checks for.

In this exercise, you should fix any issues highlighted by `Ruff` in the code below.

**Please format the code manually** as opposed to right clicking the code and formatting it automatically with "Format Notebook".

````python
def addnumbers(a, b):
    """Adds two numbers together.
    Args:
        a (int or float): The first number.
        b (int or float): The second number.
    Returns:
        int or float: The sum of the two numbers.
    """
    result = a + b
    return result


def subtract_numbers(a, b):
    """Subtracts the second number from the first number.

    Args:
        a (int or float): The first number.
        b (int or float): The second number.

    Returns:
        int or float: The result of subtracting b from a.
    """
    result = a - b
    return result


def multiplynumbers(a, b):
    """Multiplies two numbers.

    Args:
        a (int or float): The first number.
        b (int or float): The second number.

    Returns:
        int or float: The product of the two numbers.
    """
    result = a * b
    return result


def divide_numbers_but_with_a_really_long_name(a, b):
    """Divides the first number by the second number.

    Args:
        a (int or float): The numerator.
        b (int or float): The denominator.

    Returns:
        float: The result of dividing a by b.

    Raises:
        ValueError: If the denominator (b) is zero.
    """
    if b == 0:
        raise ValueError("Cannot divide by zero")
    result = a / b
    return result


def main():
    """Main function to demonstrate arithmetic operations."""
    x = 10
    y = 5
    print("Addition:", addnumbers(x, y))
    print("Subtraction:", subtract_numbers(x, y))
    print("Multiplication:", multiplynumbers(x, y))
    print("Division:", divide_numbers_but_with_a_really_long_name(x, y))


if __name__ == "__main__":
    main()
````
